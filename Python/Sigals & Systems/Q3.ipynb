```python
"""
  Signals & Systems Q3
  Author: Seyed Sobhan Hosseini
  Date: Sunday, June 25, 2023, 12:59:09 PM
"""

import numpy as np
import sympy
import matplotlib.pyplot as plt

n = np.arange(-20, 101)
delta = np.ones_like(n)
delta[n != 0] = 0
u = np.heaviside(n, 1)
y = np.zeros_like(n, float)

for i in range(2, len(n)):
  y[i] = -0.9 * y[i-2] + delta[i] + y[i-1]

h = y
plt.stem(n, h)
plt.show()

y[:] = 0
for i in range(2, len(n)):
  y[i] = -0.9 * y[i-2] + u[i] + y[i-1]

s = y
plt.stem(n, s)
plt.show()
```
