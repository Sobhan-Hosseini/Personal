```python
"""
  Signals & Systems Q2
  Author: Seyed Sobhan Hosseini
  Date: Sunday, June 25, 2023, 12:59:09 PM
"""

import matplotlib.pyplot as plt
import numpy as np

#a
n = np.arange(-5,20)
xa = (0.8**n)*np.heaviside(n,1)
plt.stem(n,xa)
plt.show()
ha = (0.8**n)*np.heaviside(n-1,1)
plt.stem(n,ha)
plt.show()
ya = np.convolve(xa,ha)
plt.stem(ya)
plt.show()
print(ya)

#b
xb = np.array([0,0,0,3,1,2])
hb = np.array([4,2,3,2])
yb = np.convolve(xb,hb)
plt.stem(yb)
plt.show()
print(yb)

#c
a = np.heaviside(n-1,1)
b = np.heaviside(n-5,1)
c = np.heaviside(n,1)
d = np.heaviside(n-4,1)
xc = np.cos(n*np.pi/4)*(a-b)
plt.stem(xc)
plt.show()
hc = np.sin(n*np.pi/4)*(c-d)*(2*n-1)
plt.stem(hc)
plt.show()
yc = np.convolve(xc,hc)
plt.stem(yc)
plt.show()
print(yc)


```
